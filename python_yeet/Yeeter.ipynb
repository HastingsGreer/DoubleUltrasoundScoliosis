```python
from PIL import ImageGrab
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import hackymouse
import time
bb = []
def setBB():
    print("move mouse to upper left hand corner")
    time.sleep(3)
    x, y = hackymouse.getPos()
    bb[:] = [x, y, x + 512, y + 512]
setBB()
print(bb)

ImageGrab.grab(bb)
```

```python
import socket

host = socket.gethostname()    
port = 9980             # The same port as used by the server
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.connect((host, port))
#s.sendall(bytes(range(256)))
#s.close()


def takeVideo(n):
    times = []
    images = []
    for _ in range(n):
        times.append(time.time())
        images.append(ImageGrab.grab(bb))
    return np.array(times), np.array([np.array(u) for u in images])

def takeShot():
    return np.array(ImageGrab.grab(bb), dtype=np.uint8)[:, :, 0]
out2 = takeShot()
s.sendall(out2.tobytes())
```
